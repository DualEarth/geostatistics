# Population and Samples, descriptive statistics

### Continuouss values

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate synthetic population: 10000 data points normally distributed
np.random.seed(101)
population = np.random.normal(loc=50, scale=15, size=100)  # Mean = 50, SD = 15

# Select a random sample from the population
sample = np.random.choice(population, size=50, replace=False)

# Descriptive statistics for the population
population_mean = np.mean(population)
population_median = np.median(population)
population_mode = stats.mode(population)[0]
population_range = np.ptp(population)
population_variance = np.var(population)
population_std_dev = np.std(population)

# Descriptive statistics for the sample
sample_mean = np.mean(sample)
sample_median = np.median(sample)
sample_mode = stats.mode(sample)[0]
sample_range = np.ptp(sample)
sample_variance = np.var(sample)
sample_std_dev = np.std(sample)

# Plotting the population data
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.hist(population, bins=30, color='skyblue', alpha=0.7, label='Population')
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(population_median, color='green', linestyle='dashed', linewidth=1, label='Median')
plt.axvline(population_mode, color='blue', linestyle='dashed', linewidth=1, label='Mode')
plt.title('Population Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Plotting the sample data
plt.subplot(1, 2, 2)
plt.hist(sample, bins=30, color='orange', alpha=0.7, label='Sample')
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(sample_median, color='green', linestyle='dashed', linewidth=1, label='Median')
plt.axvline(sample_mode, color='blue', linestyle='dashed', linewidth=1, label='Mode')
plt.title('Sample Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.show()

# Print statistics in the console
print("Population Statistics:")
print(f"Mean: {population_mean:.2f}, Median: {population_median:.2f}, Mode: {population_mode:.2f}")
print(f"Range: {population_range:.2f}, Variance: {population_variance:.2f}, Standard Deviation: {population_std_dev:.2f}\n")

print("Sample Statistics:")
print(f"Mean: {sample_mean:.2f}, Median: {sample_median:.2f}, Mode: {sample_mode:.2f}")
print(f"Range: {sample_range:.2f}, Variance: {sample_variance:.2f}, Standard Deviation: {sample_std_dev:.2f}")
```

### Discrete values

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Seed for reproducibility
np.random.seed(0)

# Generate synthetic population: Binomial distribution
# Parameters for the binomial distribution: number of trials (n) and probability of success (p)
n, p = 10, 0.5  # Example: 10 trials, 50% success rate
population = np.random.binomial(n, p, size=100)

# Select a random sample from the population
sample = np.random.choice(population, size=20, replace=False)

# Descriptive statistics for the population
population_mean = np.mean(population)
population_median = np.median(population)
population_mode = stats.mode(population)[0]
population_range = np.ptp(population)
population_variance = np.var(population)
population_std_dev = np.std(population)

# Descriptive statistics for the sample
sample_mean = np.mean(sample)
sample_median = np.median(sample)
sample_mode = stats.mode(sample)[0]
sample_range = np.ptp(sample)
sample_variance = np.var(sample)
sample_std_dev = np.std(sample)

# Plotting the population data
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.hist(population, bins=np.arange(-0.5, n+1.5), alpha=0.7, color='skyblue', label='Population', align='mid')
plt.xticks(range(n+1))
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(population_median, color='green', linestyle='dashed', linewidth=1, label='Median')
plt.axvline(population_mode, color='blue', linestyle='dashed', linewidth=1, label='Mode')
plt.title('Population Distribution (Discrete)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

# Plotting the sample data
plt.subplot(1, 2, 2)
plt.hist(sample, bins=np.arange(-0.5, n+1.5), alpha=0.7, color='orange', label='Sample', align='mid')
plt.xticks(range(n+1))
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(sample_median, color='green', linestyle='dashed', linewidth=1, label='Median')
plt.axvline(sample_mode, color='blue', linestyle='dashed', linewidth=1, label='Mode')
plt.title('Sample Distribution (Discrete)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.show()

# Print statistics in the console
print("Population Statistics:")
print(f"Mean: {population_mean:.2f}, Median: {population_median:.2f}, Mode: {population_mode:.2f}")
print(f"Range: {population_range:.2f}, Variance: {population_variance:.2f}, Standard Deviation: {population_std_dev:.2f}\n")

print("Sample Statistics:")
print(f"Mean: {sample_mean:.2f}, Median: {sample_median:.2f}, Mode: {sample_mode:.2f}")
print(f"Range: {sample_range:.2f}, Variance: {sample_variance:.2f}, Standard Deviation: {sample_std_dev:.2f}")
```
