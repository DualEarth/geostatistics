```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(0)

# Example of a categorical random variable: Types of rocks
# Let's assume 0 = Igneous, 1 = Sedimentary, 2 = Metamorphic
n_samples = 1000
rock_types = np.random.choice([0, 1, 2], size=n_samples, p=[0.5, 0.3, 0.2])
rock_type_names = ['Igneous', 'Sedimentary', 'Metamorphic']

# Example of a continuous random variable: Mineral concentration (e.g., iron % in a rock sample)
# Assuming a normal distribution for iron concentration where the mean is 20% and the std dev is 5%
iron_concentration = np.random.normal(20, 5, size=n_samples)

# Example of a discretely intervalled random variable: Mohs Hardness of minerals
# Assuming minerals are uniformly distributed across common hardness ratings from 1 to 10
mohs_hardness = np.random.randint(1, 11, size=n_samples)  # from 1 to 10

# Plotting the distribution of rock types (Categorical Random Variable)
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
counts, bins, patches = plt.hist(rock_types, bins=np.arange(-0.5, 3.5), align='mid', color='skyblue', rwidth=0.7)
plt.xticks(ticks=np.arange(3), labels=rock_type_names)
plt.title('Distribution of Rock Types')
plt.xlabel('Rock Type')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

# Plotting the distribution of mineral concentration (Continuous Random Variable)
plt.subplot(1, 3, 2)
plt.hist(iron_concentration, bins=30, color='orange', alpha=0.7)
plt.title('Distribution of Iron Concentration in Rocks')
plt.xlabel('Iron Concentration (%)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

# Plotting the distribution of Mohs Hardness (Discretely Interval Random Variable)
plt.subplot(1, 3, 3)
plt.hist(mohs_hardness, bins=np.arange(0.5, 11.5), color='green', alpha=0.7, align='mid')
plt.xticks(range(1, 11))
plt.title('Distribution of Mohs Hardness')
plt.xlabel('Mohs Hardness Scale')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

plt.tight_layout()
plt.show()
```
